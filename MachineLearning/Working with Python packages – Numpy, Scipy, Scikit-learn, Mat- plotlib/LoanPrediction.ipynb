```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Load the dataset
train = pd.read_csv('Datasets/LoanPrediction/test.csv')
test = pd.read_csv('Datasets/LoanPrediction/train.csv')

# Initial view
train.head()
```

```python
# Summary statistics
print(train.describe())

# Visualizing Loan Status (Target Variable)
plt.figure(figsize=(6,4))
sns.countplot(x='Loan_Status', data=train)
plt.title('Loan Approval Status')
plt.savefig('../Screenshots/loan_status_distribution.png') # Saves to your screenshots folder
plt.show()

# Checking for Outliers in Applicant Income using Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(x=train['ApplicantIncome'])
plt.title('Applicant Income Boxplot')
plt.savefig('../Screenshots/income_outliers.png') 
plt.show()
```
